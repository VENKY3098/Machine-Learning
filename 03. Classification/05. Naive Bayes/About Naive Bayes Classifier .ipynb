# About Naive Bayes Classsifier

There is a dataset which is having two features x1 and x2.<br>
Let X1 be salary and X2 be age,there are two classified data i.e. a portion depicts that a person walks and a second portion wich depicts that a person drives to work.

Question : If we are going to add a new feature to our dataset then where will we put this data i.e. either in walk or drive ?<br>
So the class of problem is classification.

**P(walks|X) = (P(X|walks)*P(Walks)) / P(X)**<br>
X represents feature of that datapoint. (features can be age,salary etc.)

These will be calculated in sequence of :<br>
1. P(walks) -> Prior Probaility<br>
2. P(X) -> Marginal Likelihood<br> 
3. P(X|Walks) -> Likelihood<br>
4. P(Walks|X) -> Posterior Probability

Then we will calculate P(walks|X) VS (P(X|walks)

***P(Walk|X)***

1. **P(Walks) :<br>**
P(Walks) = Number of walkers / Total Observation
P(Walks) = **10/30**

2. **P(X) :**<br>
step : We will a draw circle around our observation (radius will be decided by us).<br>
Look for all the points that are inside the circle, we are going to group them in terms of features to the point we had.Anybody who falls somewhere in that vicinity is going to be deemed similar to the new data pointwe are adding to our dataset(refer to the graph).<br><br>
How do we calclulate **Probability of X** ?<br>
The probability of X is the probability of a new point that we add to our dataset being similar in features to the point that we actually are adding to it.<br><br>
**P(X) = number of similar observations/Total Observations**<br>
P(X) = **4/30**

3. **P(X|Walks) :**<br>
Again we a draw a circle. So anything that falls inside the circle is deemed to be similar to the point that we are adding.<br>
Questtion may arise that **What is the likelihood that a randomly selected data point will be from this circle.** We will be quite biased here i.e. we will only talk about the red points. 

what is the likelihood that a randomly selected data point from our dataset is somebody who exihibit feature similar to point that we are adding to dataset.

**P(X|Walks) : Number of similar observations among those who walk / Total number of walkers**<br>
P(X|Walks) = **3/10**

```python
#P(Walk|X)
((3/10)*(10/30))/(4/30)
```

<br>**Step 2 would be same as we have done for walking instead we will find for driving.**

**P(Drive|X)**

P(Drive) = **20/30**<br>
P(X) = **4/30**<br>
P(X|Drive) = **1/20**

```python
#P(Drive|X)
((1/20)*(20/30))/(4/30)
```

**Step 3 would be to compare P(Drive|X) `VS` P(X|Walks)** <br>
**0.25 < 0.75**<br>
Therefore, **P(Drive|X) < P(X|Walks)**<br>
So it is more likely that that the person who features X is more likely to go by walk.

Therefore the new point will be in the person who walks to work.
