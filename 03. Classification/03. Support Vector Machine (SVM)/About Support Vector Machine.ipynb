# Support Vector Machine

How SVM work ?

So here we can see points on a two dimensional space which are some of the observations.

The question that arises here is **How can we derive a line that will separate these two observations i.e. the decision boundary**. It will be important if we are going to add some new points.

We can create lots of different lines which can be the decision boundary for the above classification. But they will not be **optimal** while adding a new point.

The above blue line is the line that separates the two classes and at the same time it has a maximum margin.The **blue** line is equidistant from **red** line and **green** line.

The perpendicular line between blue-red and blue-green line makes a touch with red and green line respectively which is known as support vectors.

The two points are known as support vectors because they are supporting the whole algorithm.
The other points does not contribute to the algorithm. **These point as seen in multi-dimension space act as a vector**.

The blue line in the middle is known as **Maximum Margin Hyperplane(Maximum margin classifier)**.

In 2D space it is a classifier whereas in multi-dimension it is a hyperplane.
