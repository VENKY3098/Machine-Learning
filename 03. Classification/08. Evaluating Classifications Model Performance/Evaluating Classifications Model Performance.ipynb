# False Positive and False Negative

Refer to above graph.

For observation **#1** and **#4** we predicted correctly.<br><br>
For observation **#2** and **#3** we predicted that **#2** **will not** take the offer but **it took** and for  **#3** we predicted that **it will take** the offer but it did **not** take it. 

So here we can say that our logistic regression made error.

**What does false positive mean ?**<br><br>
It means that we **predicted** a **positive outcome** but it was **false** i.e. we predicted an effect that did not occur.<br>
It is also known as **Type-1 error** 

**What does false negative mean ?**<br><br>
We predicted that there won't be an effect but the effect actually did occur. So our predicyion was negative meaning there won't be effect but it was a false.<br>
It is also known as **Type-2 error** 

***

***For Remembering purpose only***

**Type-1** is a less dangerous as it is kind of **warning** basically it is like we predicted that something is going to happen but it didn't occur.<br>
e.x. if we predicted there might occur earthquake but actually it didn't occur.<br><br>
**Type-2** is dangerous as we predict something is not going to happen but actually it does happens.
