# Hierarchial Clustering

There are two types of Hierarchical clustering <br>
1) Agglomerative<br>
2) Divisive<br>

# Agglomerative HC<br>
**Step 1** : Make each data point a single-point cluster. That forms N-Clusters.<br>
**Step 2** : Take the two closest data points and make them one cluster. That forms N-1 Clusters.<br>
**Step 3** : Take the two closest clusters and make them one cluster. That forms N-2 Clusters.<br>
**Step 4** : Repeat Step 3 until there is only one cluster.

`Closest Cluster` : While we talk about gaussian/euclidean distance we talk about the points. But clusters hold different meaning than the points. It matters how we measure distance between clusters.  

Euclidean Distance = $\sqrt{(X2-X1)^2 + (Y2-Y1)^2}$
