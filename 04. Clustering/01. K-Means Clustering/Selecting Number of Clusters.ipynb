**WCSS = within cluster sum of sqaures**

The metric to evaluate how a certain number of clusters performs compared to different number of clusters.

If there is is only one cluster then **wcss** will be a greater as each point will be measured from a single centroids, but if n = 2,there are two clusters so wcss be less than when the there was a single cluster as each distance will be of its own centroids.The more the number of centroid the lesser will be the wcss.
