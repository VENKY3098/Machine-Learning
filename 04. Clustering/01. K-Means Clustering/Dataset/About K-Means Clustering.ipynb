# K-Means Clustering

It is a algorithm that allows us to cluster your data and it is very convenient tool for discovering categories of groups in our dataset.

Question : <br>
Can we identify certain groups among all variables ?

K-Means takes out the complexity from this decision making process and  allows you to very easily identify those clusters.

**steps**<br>
***

**Step 1** : Choose the number **K** of clusters.<br>
***
**Step 2** : Select at random K points, the centroids(not necessarily from your dataset).<br> 
We can select any point on the scatterplot, they don't have to be part of the observations that are being plotted on the scatter-plot.<br>
***
**Step 3** : Assign each data point to the closest centroid => That forms K clusters.<br>
There is going to be an iterative process to refine those clusters.<br>
Check for every point in a dataset which out of the centroid that we have identified is closest to the specific data-point.<br>
***
**Step 4** : Compute and place the new centroid of each cluster.<br>
***
**Step 5** : Reassign each data point to the new closest centroid. If any reassignment took place, go to step 4, otherwise go to finish.

We can scatter-plot when when there are two features but we cannot plot a scatter-plot if there are more than two features as it will be complex. And that is where k-means clustering comes into play.
