# Gradient Descent and Stochastic Gradient Descent

Why can't we follow bruteforce method ?<br>
For large datasets it will be very ineffective method to follow bruteforce approach.<br>
For gradient descent the cost function should be convex i.e it should have a global minimum.<br>
For example if the cost function is not square difference then the graph will not be convex.<br>
So rather than finding the global minimum we could find a local minimum. So we need a gradient descent named as stochastic gradient descent.<br>
In gradient Descent we adjust the weight after running the whole row being plugged in the neural network at the same time but in Stochastic gradient descent we plug each row at a time in neural network after which we accordingly adjust the weight.

# Backpropagation

There is a process called Forward Propagation where information is entered  into the input layer and then it is propagated forward to get our $\hat{y}$, our output values and then we compare those to the actual values that we have in our training set and then we calculate the errors then the error gets backpropagated through the network in opposite direction and that allows us to train the network by adjusting the weight. Backpropagation allows us to adjust the weight simultaneouly.<br>
During the process of backpropagation simply because of the weight, the algorithm is structred in such a way that we are able to we are able to adjust the weight at the same time. So we basically know which part of the error each of weights in the Neural Network is responsible for. 

# Training the ANN with Stochastic Gradient Descent

**Step 1 :** Randomly inintialise the weights to small numbers close to 0 (but not 0).<br>
**Step 2 :** Input the First observation of your dataset in the input layer,each feature in one input node.<br>
**Step 3 :** `Forward-Propagation` : From left to right, the neurons are activated in a way that the impact of each neuron's activation is limited by the weights. Propagate the activations until getting the predicted result y.<br>
**Step 4 :** Compare the predicted result to the actualresult. Measure the generated error.<br>
**Step 5 :** `Back-propagation` From right to left, the error is back-propagated. Update the weights according
to how much they are responsible for the error. The learning rate decides by how much we update the weights.<br>
**Step 6 :** Repeat Steps 1 to 5 and update the weights after each observation (Reinforcement Learnning) or Repeat Steps 1 to 5 but update the weights only after a batch of observations(Batch Learning).<br>
**Step 7 :** When the whole training set passed through the ANN, that makes an epoch. Redo more epochs.
