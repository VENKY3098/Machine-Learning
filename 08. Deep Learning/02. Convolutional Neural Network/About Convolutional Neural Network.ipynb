# Step 1 - Convolution

**Convolution Function :**<br>
$$ (f\ *\ g)(t)\stackrel{def}{=}\ \int_{-\infty}^{\infty}f(\tau)g(t-\tau)d\tau $$<br>
A convolution is basically a combined integration of two function and it shows how one function modifies the other or modifies the shape of the other.

So we apply rectifier function is because we want to increase non-linearity in our image or in our network. We want to increase non-linearity in our network is because images themselves are highly non-linear especially if we are recogonizing different objects.<br>

# Step 2 -Pooling

In pooling process we take a n\*n selected matrix and then from the stride we select the maximum number.<br>
Pooling is also known as **Downsampling**.
