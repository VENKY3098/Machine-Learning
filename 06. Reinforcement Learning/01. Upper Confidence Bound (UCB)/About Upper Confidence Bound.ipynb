# Upper Confidence Bound

The problem we are solving is a multi-armed bandit problem where we have got many slot machines and we can bet our money in any one of them and we need to find out how  to bet to **maximize our return**. <br>
Another problem similar to this is that we see multiple ads of certain thing and we need to find which one of the best out of it. 

1) We have d arms. For example, arms are ads that we display to users each time they connect a web page.<br>
2) Each time a user connects to this web page, that makes a round.<br>
3) At each round n, we choose an ad to display to the user.<br>
4) At each round n, ad gives reward r$_{i}$(n) $\epsilon$ {0,1}: r$_{i}$(n)=1 if the user clicked on the ad i, 0 if the user didn't.<br>
5) Our goal is to maximize the total reward we get over many rounds.

**Step 1 :** At each round n, we consider two numbers for each ad i:<br>
* N$_{i}$(n) - the number of times the ad i was selected up to round n,<br>
* R$_{i}$(n) - the sum of rewards of the ad i up to round n.

**Step 2 :** From these two numbers we compute :<br>
* the average reward of ad i up to round n<br>
$\overline{r}$(n) = $\frac{R_{i}(n)}{N_{i}(n)}$<br>
* The confidence interval [$\overline{r}_{i}$(n) - $\Delta_{i}$(n),$\overline{r_{i}}$(n) + $\Delta{_{i}}$(n)] at round n with <br>
$$\Delta_{i}(n) = \sqrt{\frac{3\log(n)}{2N_{i}(n)}}$$

**Step 3 :** we select the ad $i$ that has the maximum UCB $\overline{r_{i}}(n)+\Delta_{i}(n)$ 

If we plot normal distribution for different observations then at last we can tell which will better for betting.<br>
But our need is to know which machine gives best outcome while it is in the process.<br><br>
How does the algorithm works ?<br>
So we start with putting mean value on the vertical axis, then it **assumes certain starting point**.<br>
Then the formula behind the algorithm create a `confidence Band`, It is designed in such a way that a high level of certainity that confidence band will include the actual expected return.<br> Basically the first couple of rounds are going to be trial runs Son we going to intensionally try out machines at least one time each in order for us to be able to place the start value at its place and come up with a confidence band.<br>
How do we pick the machine ?<br>
We pick the machine with `highest Confidence bound`, we talk generally about the upper confidence bound.<br>
Randomly pick any of the box, for instance take the case of ad, if the person didn't click the ad the start red value(observed average) shifts downward for that particular machine.<br>
In the long run observed average is going to converge to the expected value for that given distribution. So Confidence bound/interval becomes smaller.<br>
Next step is to find next one with highest confidence bound.
The average will be affected by the click of the person. And hence it will converge with the expected value. The second thing that happens with the bound is that because of an `additional observation point in sample` now we are now more confident in this interval.<br>
`The confidence bounds sole purpose is to include the actual expected value wherever it is.` The middle line is somewhere between the bounds and due to addition of new sample data the  confidence bound decreases(boundary of the box). <br>
If the decision is correct then the average value goes near the expected value and vice-versa. I a new sample point is added then the confidence-bound decreases.
