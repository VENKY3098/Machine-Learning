# Question
* How do i know which model to choose for my machine learning ?<br>
So if we have a machine learning problem that comes with a specific dataset,<br>
* Step-1 we need to know if our problem is a regression problem, classification problem or a clustering problem.<br>
If we don't have dependent variable then it is a **clustering problem**,<br>
And if we have a dependent variable then we see if it has a continuous outcome or categorical outcome,<br>
If it's a **continuous outcome** then outcome then our problem is  a **regression problem** and if it's a **categorical outcome** then our problem is a **classification problem**.<br>
Next we need to know wether our problem is **linear problem** or a **non-linear problem** ?<br>

# K-Fold Cross Validation

`Validation Method` fixes the variance problem with the splitting of datasets.<br>
**So how does the validation method work ?**<br>
Validation Method splits the dataset into two parts i.e. training set and test set.<br>
Now the training set is splitted into 10 folds i.e. when $k = 10$, now we train our model on 9 folds and test it on the 1 fold.  

```python
Image(filename="Images\image1.jpg",height=500,width=500)
```

So after that we can take average of the different accuracies of the 10 evaluations and also compute the standard deviation to have a look at the variance.  

```python
Image(filename="Images\image2.jpg",height=500,width=500)
```

So after that we can know in which of these four categories our model will lie. If we get a `good accuracy` and a `small variance` it is the lower left graph, if we get a `large accuracy` and `high variance` it will be on lower right one, if we get `small accuracy` and `low variance` it is on upper left and if we get `low accuracy` and `high variance` we will be on upper right.
