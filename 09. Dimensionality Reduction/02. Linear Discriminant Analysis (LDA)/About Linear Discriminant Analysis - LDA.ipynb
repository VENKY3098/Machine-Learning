# About Linear Discriminant Analysis - LDA
It shares some similarity with Principal Component Analysis.<br>
* Used in the pre-processing step for pattern classification
* Has the goal to project a dataset onto a lower-dimensional space
***
How is LDA different from PCA ?<br>
LDA differs from PCA because in `addition to finding the component axises with LDA` we are interested in the `axes that minimizes the separation between multiple classes`.<br>
More precisely : <br>
The goal of LDA is to project a feature space (a dataset $n-$dimensional samples) onto a small subspace $k\ (where\ k\leq n-1)$ while maintaining the class-discriminatory information.<br>
Both PCA and LDA are linear transform techniques used for dimensional reduction. PCA is described as unsupervised but LDA is supervised because of the relation to the dependent variable.

# Steps
1. Compute the d-dimensional mean vectors for the different classes from the datasets.
2. Compute the Scatter matrices (in-between-class and within-class scatter matrix).
3. Compute the eigenvectors $(e_1,\ e_2,\ .\ .\ .\ ,\ e_d )$ and corresponding eigenvalues $(\lambda_1,\ \lambda_2,\ .\ .\ .\ \lambda_d)$ for the scatter matrices.
4. Sort the eigenvectors by decreasing eigenvalues to form a $d\ x\ k$ dimensional matrix $\textbf{W}$(Where every column represents an eigenvector.
5. use this $d\ x\ k$ eigenvector matrix to transform the samples onto the new subspace. This can be summariszed by the matrix multiplication: $\textbf{Y}\ =\ \textbf{X}\ x\ \textbf{W}$(Where $\textbf{X}$ is a nxd d-dimensional matrix representing the $\textit{n}$ samples,and $\textit{y}$ are the transformed $n\ x\ k$- dimensional samples in the new subspace).

**LDA in Short**<br>
From the n independent variables of your dataset ,LDA extracts p $\leq$ n new independent variables that separate the most the classes of dependent variable.<br>
